```python
import requests 
import StringIO
import zipfile
import numpy as np
import pandas as pd

%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
fulldata = pd.read_csv('weathersentiment.csv')
fulldata.drop_duplicates(subset='tid', inplace=True)
fulldata = fulldata.drop(fulldata.columns[0], axis=1)
tws = fulldata[fulldata.polarity != 0]
tws.head()
```

# Added max alpha to Ever's scatter plots to see if there was a relationship:

```python
plt.scatter(tws.polarity, tws.TemperatureF, alpha = 0.002)
plt.title('Sentiment vs Temperature Across All Cities')
plt.xlabel('Sentiment')
plt.ylabel('Temperature')
plt.show()
```

### Still too dense of data, even with max alpha! Went with a 100,000 item random sample:

```python
import random
rows = random.sample(tws.index, 100000)
rand = tws.ix[rows]
```

```python
plt.scatter(rand.polarity, rand.TemperatureF, alpha = .005)
plt.title('Sentiment vs Temperature')
plt.xlabel('Sentiment')
plt.ylabel('Temperature')
```

```python
plt.scatter(rand.polarity, rand.Humidity, alpha = .005)
plt.title('Sentiment vs Humidity')
plt.xlabel('Sentiment')
plt.ylabel('Humidity')
```

```python
plt.scatter(rand.polarity, rand.SeaLevelPressureIn, alpha = .002)
plt.title('Sentiment vs SeaLevelPressureIn')
plt.xlabel('Sentiment')
plt.ylabel('SeaLevelPressureIn')
```

```python
plt.scatter(rand.polarity, rand.PrecipitationIn, alpha = .01)
plt.title('Sentiment vs PrecipitationIn')
plt.xlabel('Sentiment')
plt.ylabel('PrecipitationIn')
plt.ylim(-.05,.2)
```

# Plotting of Summary Statistics

```python
fullcities = fulldata.groupby('City')
```

```python
citymean = fullcities.mean()
```

```python
citymean.index
```

```python
fig, ax = plt.subplots()

N = 7
ind = np.arange(N)
width = .4

rects2 = ax.bar(ind, citymean.Humidity)
ax.set_xticks(ind + width)
ax.set_xticklabels(('bos', 'chi', 'la', 'lon', 'mia', 'ny', 'sea'))
plt.title("Humidity by City")

plt.show()
```

```python
fig, ax = plt.subplots()

N = 7
ind = np.arange(N)
width = .4

rects2 = ax.bar(ind, citymean.TemperatureF, color = "red")
ax.set_xticks(ind + width)
ax.set_xticklabels(('bos', 'chi', 'la', 'lon', 'mia', 'ny', 'sea'))
plt.title("TemperatureF by City")

plt.show()
```

```python
fig, ax = plt.subplots()

N = 7
ind = np.arange(N)
width = .4

rects2 = ax.bar(ind, citymean.PrecipitationIn, color = "cyan")
ax.set_xticks(ind + width)
ax.set_xticklabels(('bos', 'chi', 'la', 'lon', 'mia', 'ny', 'sea'))
plt.title("PrecipitationIn by City")

plt.show()
```
