```python
import requests 
import StringIO
import zipfile
import numpy as np
import pandas as pd

%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
tweet = pd.read_csv('weathersentiment.csv')
```

```python
# gets rid of random Unnamed:0 column

tweet = tweet.drop(tweet.columns[0], axis=1)
```

```python
tweet.drop_duplicates(subset='tid', inplace=True)
```

```python
tweet = tweet[tweet.polarity != 0]
```

```python
# quick statistics

cities = tweet.groupby('City')
```

```python
cities.mean()
```

```python
cities.median()
```

It looks like while the weather varies by a lot across cities (as expected), average sentiment is very similar and median sentiment is the same!

```python
# total relationship between sentiment and temperature

plt.scatter(tweet.polarity, tweet.TemperatureF)
plt.title('Sentiment vs Temperature Across All Cities')
plt.xlabel('Sentiment')
plt.ylabel('Temperature')
plt.show()
```

Overall, no relationship between temperature and sentiment. Next will look at sentiment and other factors.

```python
# http://stackoverflow.com/questions/12190874/pandas-sampling-a-dataframe
# takes random sample of tweets
# makes plotting a lot easier

import random

rows = random.sample(tweet.index, 10000)
rand = tweet.ix[rows]
```

```python
plt.scatter(rand.polarity, rand.TemperatureF)
plt.title('Sentiment vs Temperature')
plt.xlabel('Sentiment')
plt.ylabel('Temperature')
```

```python
plt.scatter(rand.polarity, rand.Humidity)
plt.title('Sentiment vs Humidity')
plt.xlabel('Sentiment')
plt.ylabel('Humidity')
```

```python
plt.scatter(rand.polarity, rand.SeaLevelPressureIn)
plt.title('Sentiment vs SeaLevelPressureIn')
plt.xlabel('Sentiment')
plt.ylabel('SeaLevelPressureIn')
```

```python
plt.scatter(rand.polarity, rand.PrecipitationIn)
plt.title('Sentiment vs PrecipitationIn')
plt.xlabel('Sentiment')
plt.ylabel('PrecipitationIn')
```

There appears to be no identifiable relationship between any of the factors and sentiment. Next, I will isolate by city.

Also, above, I randomly sampled 10K tweets to make the graphs. 10K is arbitrary but allows for faster run-time.
