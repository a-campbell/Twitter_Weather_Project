# Scraping Weather Data

At this point, we had gathered 4.5 million tweets and were ready to find relevant weather data. We built this weather scraper, which uses the data available at wunderground in order to generate a dataframe with hourly weather data for each city in our analysis. Users of this scraper are able to choose a customized start and end date for their analysis, which is easily editable via the "startdate" and "enddate" variables. For our purposes, we matched the date range of weather data with the date range that our tweet scraper was running (November 21 through December 2)

```python
# Copying code for getting weather data for relevant dates
from StringIO import StringIO
import datetime as dt
import time
import bs4
import os
import sys
import requests
# choose a start and end date
startdate = dt.date(2014, 11, 21)
enddate = dt.date(2014, 12, 2)

# preparation stuff
current = startdate
allweatherdata = pd.DataFrame(columns = ['City', 'TemperatureF','Humidity','SeaLevelPressureIn','WindMPH','PrecipitationIn','Events','Conditions','DateUTC'])

# iterate through weather data by date
while current <= enddate:
    urltime =  str(current.year) + "/" + str(current.month) + "/" + str(current.day)
    current += dt.timedelta(days=1)
    
    # generate list of city urls with custom dates
    mia = "http://www.wunderground.com/history/airport/KMIA/" + urltime + "/DailyHistory.html?req_city=Miami&req_state=FL&req_statename=Florida&MR=1&format=1"
    sea = "http://www.wunderground.com/history/airport/KBFI/" + urltime + "/DailyHistory.html?req_city=Seattle&req_state=WA&req_statename=Washington&MR=1&format=1"
    ny = "http://www.wunderground.com/history/airport/KNYC/" + urltime + "/DailyHistory.html?req_city=New+York&req_state=NY&req_statename=New+York&MR=1&format=1"
    la = "http://www.wunderground.com/history/airport/KCQT/" + urltime + "/DailyHistory.html?req_city=Los+Angeles&req_state=CA&req_statename=California&MR=1&format=1"
    lon = "http://www.wunderground.com/history/airport/EGLL/" + urltime + "/DailyHistory.html?req_city=London&req_state=&req_statename=United+Kingdom&format=1"
    bos = "http://www.wunderground.com/history/airport/KBOS/" + urltime + "/DailyHistory.html?req_city=Boston&req_state=MA&req_statename=Massachusetts&MR=1&format=1"
    chi = "http://www.wunderground.com/history/airport/KMDW/" + urltime + "/DailyHistory.html?req_city=Chicago&req_state=IL&req_statename=Illinois&MR=1&format=1"
    cities = [mia, sea, ny, la, lon, bos, chi]
    citynames = ["mia", "sea", "ny", "la", "lon", "bos", "chi"]
    
    # iterate through weather data by city
    for city in range(0, len(cities)):
        
        url = cities[city]
        citycol = []

        weatherdata = requests.get(url)
        weatherdata = weatherdata.content
        weatherdata = pd.read_csv(StringIO(weatherdata), header=1)
        weatherdata.rename(columns={"DateUTC<br />":"DateUTC", "Sea Level PressureIn":"SeaLevelPressureIn", "Wind SpeedMPH":"WindMPH"}, inplace=True)

        # get rid of </br> from DateUTC weatherdata 
        for i in range(0, len(weatherdata["DateUTC"])):
            weatherdata["DateUTC"][i] = weatherdata["DateUTC"][i][:19]
            
        # convert DateUTC to datetime and only keep hourly information and add values to a series that will be a city column
        weatherdata["DateUTC"] = pd.to_datetime(weatherdata.DateUTC)
        mcount = len(weatherdata['DateUTC'])
        for i in range(0, len(weatherdata["DateUTC"])):
            if (24*(i+1))%((mcount/24)+1) != 0:
                weatherdata["DateUTC"][i] = 0
            else:
                citycol.append(citynames[city])
                weatherdata["DateUTC"][i] -= dt.timedelta(minutes = weatherdata["DateUTC"][i].minute, seconds = weatherdata["DateUTC"][i].second)
                weatherdata["DateUTC"][i] += dt.timedelta(hours = 1)
        weatherdata = weatherdata[weatherdata.DateUTC != 0]

        # clean unwanted columns
        weatherdata = weatherdata.drop(['Dew PointF', 'VisibilityMPH', 'Wind Direction', 'Gust SpeedMPH', 'WindDirDegrees'], axis=1)
        weatherdata = weatherdata.drop(weatherdata.columns[0], axis=1)
        
        # add city column and append data for this city to the full data set
        weatherdata["City"] = citycol        
        allweatherdata = allweatherdata.append(weatherdata)
```

# Prepare Data for Weather/Tweet Merge

In order to Merge the Weather and Tweet Data, we needed to do a few things:
- Match column names and data types in each DataFrame for columns that need to be merged
- Round time values to the nearest hour in the tweets data set (so that it will match hourly weather values)
- Calculate a city for each tweet based on latitude and longitude (we did this by finding the nearest city based on a dictionary of city name keys and their coordinates)
- Merge using a left merge into the tweets dataframe (adding weather to twitter data)

```python
# convert tweet time to datetime and match names of columns
tweets['City'] = range(len(tweets))
tweets = tweets.rename(columns = {'time':'DateUTC'})
tweets.DateUTC = pd.to_datetime(tweets.DateUTC)
```

```python
# round tweet datetime to hourly values in the same was as weather data
for i in range(0, len(tweets.DateUTC)):
    tweets.DateUTC[i] -= dt.timedelta(minutes = tweets.DateUTC[i].minute, seconds = tweets.DateUTC[i].second)
    tweets.DateUTC[i] += dt.timedelta(hours = 1)
```

```python
# city names with lon/lat
citylatlon = {'mia': [25.761680, -80.191790], 'sea':[47.606209, -122.332071], 'ny':[40.712784, -74.005941], 'la':[34.052234, -118.243685], 'lon': [51.507351,-0.127758] , 'bos': [42.358486, -71.060097], 'chi':[41.878114, -87.629798] }
```

```python
# find closest city for tweets

for i in range(0, len(tweets.lon)):
    diff = 1000
    city = ""
    for key in citylatlon:
        curr = abs(citylatlon[key][1] - tweets.lon[i]) + abs(citylatlon[key][0] - tweets.lat[i])
        if curr < diff:
            diff = curr
            city = key
    if diff == 1000:
        city = "NaN"
    tweets['City'][i] = city
```

```python
# merge weather data onto tweet data based on city then date
weathersentiment = pd.merge(left = tweets, right = allweatherdata, how = 'inner', on = ['City', 'DateUTC'])
```

These graphs are a good first step, but they don't really give us much clear insight.
We can really group the conditions tags into two groups: rainy and clear. 

```python
#Rainy and clear polarity by city
fig,ax = plt.subplots(4, 2, figsize=(20,23))
x = 0
y = 0

for i in citynames:
    xx = tw[tw.City == i][tw.Conditions.isin(['Rain', 'Drizzle', 'Rain Showers', 'Light Snow', 'Heavy Rain', 'Heavy Drizzle', 'Mist', 'Light Rain','Ice Pellets', 'Small Hail', 'Light Ice Pellets'])].polarity
    yy = tw[tw.City == i][tw.Conditions.isin(['Clear', 'Mostly Cloudy', 'Partly Cloudy', 'Scattered Clouds'])].polarity
    ax[y,x].hist(xx.values, bins=8, alpha=0.5, label='Rainy', normed=True)
    ax[y,x].hist(yy.values, bins=8, alpha=0.5, label='Clear', normed=True)
    ax[y,x].legend(loc='upper left')
    ax[y,x].set_title(i)
    ax[y,x].set_xlabel('Polarity', size="small")
    x = x+1
    if x == 2:
        x = 0
        y = y + 1
fig.text(0.5, 0.94, 'Distribution of Tweet Polarity on Rainy vs. Clear Hours', ha='center', va='center', size="xx-large")
```

Looking at the differences in these distributions, we can see that twitter users in Chicago, LA, and Seattle appear to show some inclination towards happier tweets on clear hours and more negative tweets in rainy hours. 

It is plausible that weather will have a more significant impact on the mood of twitter users during the day (when they are more likely to be outside). To test this we can repeat the previous analyis on only daytime tweets. 

```python
#drop tweets from between 7PM-5AM
daytime = tw[tw.Hours < 19]
daytime.reset_index(inplace=True)
daytime = daytime[daytime.Hours > 5]
```

```python
#Now lets run the weather sentiment graphs again... 
fig,ax = plt.subplots(4, 2, figsize=(20,23))
x = 0
y = 0

for i in citynames:
    xx = daytime[daytime.City == i][daytime.Conditions.isin(['Rain', 'Drizzle', 'Rain Showers', 'Light Snow', 'Heavy Rain', 'Heavy Drizzle', 'Mist', 'Light Rain','Ice Pellets', 'Small Hail', 'Light Ice Pellets'])].polarity
    yy = daytime[daytime.City == i][daytime.Conditions.isin(['Clear', 'Mostly Cloudy', 'Partly Cloudy', 'Scattered Clouds'])].polarity
    ax[y,x].hist(xx.values, bins=8, alpha=0.5, label='Rainy', normed=True)
    ax[y,x].hist(yy.values, bins=8, alpha=0.5, label='Clear', normed=True)
    ax[y,x].legend(loc='upper left')
    ax[y,x].set_title(i)
    ax[y,x].set_xlabel('Polarity', size="small")
    x = x+1   
    if x == 2:
        x = 0
        y = y + 1
fig.text(0.5, 0.94, 'Distribution of Tweet Polarity on Rainy vs. Clear Daytime Hours', ha='center', va='center', size="x-large")
```

These look a little different! Chicago looks much less in favor of clear hour than before. It looks like daytime tweeters in New York actually prefer rain! Los Angeles tweeters still shows a decently strong preference for clear hours. 

What if we use a precipitation cuttoff as a proxy for rainy and sunny?

```python
#Create cuttoff of .02 inches per hour
rainy = daytime[daytime.PrecipitationIn > 0.02]
notrainy = daytime[daytime.PrecipitationIn < 0.02]

fig,ax = plt.subplots(3, 2, figsize=(20,23))
x = 0
y = 0

#Unfortunately something happned to our London precipitation data and we must exclude it. 
for i in citynames:
    if i != 'lon':
        xx = rainy[rainy.City == i].polarity
        yy = notrainy[notrainy.City == i].polarity
        ax[y,x].hist(xx.values, bins=8, alpha=0.5, label='Very Rainy', normed=True)
        ax[y,x].hist(yy.values, bins=8, alpha=0.5, label='Not Rainy', normed=True)
        ax[y,x].legend(loc='upper left')
        ax[y,x].set_title(i)
        ax[y,x].set_xlabel('Polarity', size="small")
        x = x+1
        if x == 2:
            x = 0
            y = y + 1
fig.text(0.5, 0.94, 'Distribution of Tweet Polarity on Daytime Very Rainy vs. Not Very Rainy Hours', ha='center', va='center', size="x-large")
```

It is interesting that these distribution are different from those we saw with "Conditions" tags. It appears that New York, Seattle, and Los Angeles tweeters are substantially more happy when it is not raining hard.

# Conclusion


In the end, we found that there is no clear cut relationship between weather and Twitter sentiment in the cities we observed based on our classifier for Tweet sentiment. Depending on how we classified weather conditions, different cities appeared to seek become more or less happy in either weather grouping. In general, Los Angeles, Seattle, and Chicago based users were more likely to show the relationship we expected; namely, more positive tweets during sunny weather versus rainy weather. New Yorkers are significantly less happy when it is raining heavily but appear to prefer cloudy days. 

We did find several other interesting relationships in the data:

- The majority of Tweets do not show high polarity as defined by our classifier. This problem will need to be solved by future researchers looking to investigate the relationship between Twitter Sentiment and weather data.
-  Of the cities we investigated, those with high precipitation had moderate temperatures and humidity (Boston and New York). The two driest cities had the highest temperatures (LA and Miami).
-  People seem to tweet the most late at night, with a noticeable spike around 10pm.
- Sentiment follows very noticeable patterns depending on the time of day. Additionally, time seems to be a factor for temperature and humidity but not rainfall. 
- Americans seem to be happiest around 10 or 11 in the morning, and get progressively less happy throughout the day.
- New York and L.A. produce more geo-tagged tweets than other cities.
