```python
from textblob import TextBlob
import requests
import json
import pandas as pd
```

```python
# example sentiment analysis with TextBlob

pos = 'This is the best day of my life!'
neg = 'I hate everything!'

# positive 
sentp = TextBlob(pos)
rating = sentp.sentiment.polarity
print 'Sentiment for' , sentp , 'is ' , rating

# negative
sentn = TextBlob(neg)
rating2 = sentn.sentiment.polarity
print 'Sentiment for' , sentn, 'is ', rating2
```

```python
# working with JSON

url = 'https://raw.githubusercontent.com/theandycamps/Twitter_Weather_Project/master/tweetoutput.json'

tweets = requests.get(url)

lines = tweets.text.split('\r\n')
```

```python
# first Tweet
# Tweet info structure example

example = json.loads(lines[0])
example
```

```python
# extracting relevant data

time = []
user = []
text = []
for i in xrange(0, 6290):
    ex = json.loads(lines[i])
    time.append(ex['created_at'])
    user.append(ex['user']['screen_name']) 
    text.append(ex['text'])
```

```python
# reading in to DataFrame

master = pd.DataFrame()

master['user'] = user
master['time'] = time
master['text'] = text
```

```python
master.head()
```

```python
# calculating sentiment using TextBlob

rate = []

for i in xrange(0, 6290):
    
    hold = master['text'][i]
    sho = TextBlob(hold)
    pos = sho.sentiment.polarity
    rate.append(pos)
```

```python
# adding sentiment info

master['sentiment'] = rate
master.head()
```
